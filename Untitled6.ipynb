```python
import pickle
import random
def addnew():
    f=open("apparatus management.dat","ab")

    while True:

        pname=input("ENTER THE NAME OF THE PASSENGER:")
        page=int(input("ENTER AGE:"))
        jdate=input("ENTER DATE(dd/mm/yyyy):")
        jfrom=input("JOURNEY FROM:")
        jto=input("JOURNEY TO:")

        ticketno=random.randint(1000,10000)
        data=(pname,page,jdate,jfrom,jto,ticketno)
        pickle.dump(data,f)


        ch=input("Y->YesN->No Do you want to add more :")
        if ch=='n' or ch=='N':
            break


    f.close()


def search_pname():


    ch=0
    try:
        while True:

            a=input("ENTER THE NAME OF THE PASSENGER:")
            f=open("apparatus management.dat","rb")
            r=pickle.load(f)
            if r[0]==a:
                ch=ch+1
                print("PASSENGER NAME:",r[0])
                print("AGE:",r[1])
                print("DATE:",r[2])
                print("JOURNEY FROM:",r[3])
                print("TO:",r[4])
                print("TICKET NO.:",r[5])



                kp=input("Y->Yes\nN->No\n Do you want to search more :")
                if kp=='n' or kp=='N':
                    break
    except:
        if ch==0:
            print("PASSENGER NOT FOUND.")
    f.close()




def search_age():

    t=0
    try:
        while True:

            a=input("ENTER AGE:")
            f=open("apparatus management.dat","rb")
            r=pickle.load(f)
            if r[1]==a:
                t=t+1
                print("PASSENGER NAME:",r[0])
                print("AGE:",r[1])
                print("DATE:",r[2])
                print("JOURNEY FROM:",r[3])
                print("TO:",r[4])
                print("TICKET NO.:",r[5])


                nu=input("Y->Yes\nN->No\n Do you want to search more :")
                if nu=='n' or nu=='N':
                    break
    except:
        if t==False:
            print("AGE Not Found")
    f.close()


def search_date():

    m=0
    try:
        while True:

            a=input("ENTER DATE:")
            f=open("apparatus management.dat","rb")
            r=pickle.load(f)
            if r[2]==a:
                m=m+1
                print("PASSENGER NAME:",r[0])
                print("tAGE:",r[1])
                print("DATE:",r[2])
                print("JOURNEY FROM:",r[3])
                print("TO:",r[4])
                print("TICKET NO.:",r[5])
                ball=input("Y->Yes\nN->No\n Do you want to search more :")
                if ball=='n' or ball=='N':
                    break
    except:
        if m==0:
            print(" No RESERVATION OF This Day")
    f.close()

def search_jfrom():

    cj=0
    try:
        while True:

            a=int(input("JOURNEY FROM:"))
            f=open("apparatus management.dat","rb")
            r=pickle.load(f)
            if r[3]==a:
                cj=cj+1
                print("PASSENGER NAME:",r[0])
                print("AGE:",r[1])
                print("DATE:",r[2])
                print("JOURNEY FROM:",r[3])
                print("TO:",r[4])
                print("TICKET NO.:",r[5])


                pop=input("Y->Yes N->No Do you want to search more :")
                if pop=='n' or pop=='N':
                    break
    except:
        if cj==0:
            print("CITY NOT FOUND")
    f.close()



def search_jto():

    sh=0
    try:
        while True:

            a=int(input("DESTINATION:"))
            f=open("apparatus management.dat","rb")
            r=pickle.load(f)
            if r[4]==a:
                sh=sh+1
                print("PASSENGER NAME:",r[0])
                print("tAGE:",r[1])
                print("DATE:",r[2])
                print("JOURNEY FROM:",r[3])
                print("TO:",r[4])
                print("TICKET NO.:",r[5])


                pop=input("Y->YestN->No Do you want to search more :")
                if pop=='n' or pop=='N':
                    break
    except:
        if sh==0:
            print("CITY NOT FOUND")
    f.close()


def search_tno():

    cm=0
    try:
        while True:

            a=int(input("TICKET NO.:"))
            f=open("apparatus management.dat","rb")
            r=pickle.load(f)
            if r[5]==a:
                cm=cm+1
                print("PASSENGER NAME:",r[0])
                print("AGE:",r[1])
                print("DATE:",r[2])
                print("JOURNEY FROM:",r[3])
                print("TO:",r[4])
                print("TICKET NO.:",r[5])


                po=input("Y->Yes N->No\n Do you want to search more :")
                if po=='n' or po=='N':
                    break
    except:
        if cm==0:
            print("TICKET NO. NOT FOUND.")
    f.close()

def displayall():

    f=open("apparatus management.dat","rb")
    try:
        while True:

            jp=pickle.load(f)
            print("PASSENGER NAME:",jp[0])
            print("AGE:",jp[1])
            print("DATE:",jp[2])
            print("JOURNEY FROM:",jp[3])
            print("TO:",jp[4])
            print("TICKET NO.:",jp[5])

    except:
        f.close()
        print("All Data Displayed Above")


def menu_driven():
    print("RAILWAY RESERVATION SYSTEM")
    while True:
        print("1.ENTER YOUR DETAILS t2.SEARCH BY PASSENGER NAME t3.SEARCH BY AGE t4.SEARCH BY DATE 5.SEARCH BY CITY WHERE THE JOURNEY START t6.SEARCH BY DESTINATION 7.SEARCH BY TICKET NO. t8.DISPLAY ALL TICKETS")
        choice=int(input("CHOOSE FROM ABOVE:"))
        if choice==1:
            addnew()
        elif choice==2:
            search_pname()
        elif choice==3:
            search_age()
        elif choice==4:
            search_date()
        elif choice==5:
            search_jfrom()
        elif choice==6:
            search_jto()
        elif choice==7:
            search_tno()
        elif choice==8:
            displayall()

        else:
            print("INVALID CHOICE ")
        ans=input("Y->YesN->No Do you want to continue :")
        if ans=='n' or ans=='N':
            break
    print("THANK YOU")
menu_driven()
```
